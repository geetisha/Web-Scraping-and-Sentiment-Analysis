```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time

# Set the path to chromedriver (Update this path)
CHROMEDRIVER_PATH = r"C:\Users\Geetisha Mishra\Documents\Projects\Axion Ray\chromedriver.exe"

# Setup WebDriver
service = Service(CHROMEDRIVER_PATH)
driver = webdriver.Chrome(service=service)

# Open the BestBuy reviews page
url = "https://www.bestbuy.ca/en-ca/product/microsoft-surface-laptop-13-8-touchscreen-copilot-pc-laptop-snapdragon-x-plus-16gb-ram-512gb-ssd-2024-black-en/17924130/review"
driver.get(url)

# Wait until the reviews load
wait = WebDriverWait(driver, 15)

# Function to load all reviews by clicking "Show More"
def load_all_reviews():
    while True:
        try:
            show_more_button = wait.until(EC.element_to_be_clickable((By.XPATH, "//button[contains(text(), 'Show More')]")))
            driver.execute_script("arguments[0].click();", show_more_button)
            time.sleep(2)  # Allow time for new reviews to load
        except:
            break  # No more "Show More" button, exit loop

# Load all reviews
print("Loading all reviews...")
load_all_reviews()
time.sleep(3)
print("All reviews loaded!")

# Extract all reviews
reviews = []
review_elements = driver.find_elements(By.XPATH, "//li[contains(@class, 'review')]")

print(f"Total reviews found: {len(review_elements)}")

for review in review_elements:
    try:
        title = review.find_element(By.XPATH, ".//div[contains(@class, 'reviewTitle')]//span").text
    except:
        title = "Not Available"

    try:
        text = review.find_element(By.XPATH, ".//div[contains(@class, 'reviewContent')]").text
    except:
        text = "Not Available"

    try:
        date = review.find_element(By.XPATH, ".//span[contains(@class, 'locationAndTime')]").text
    except:
        date = "Not Available"

    try:
        rating_text = review.find_element(By.XPATH, ".//p[contains(text(), 'Customer rating')]").text
        rating = rating_text.split(":")[1].split(" out of")[0].strip()  # Extracts rating number
    except:
        rating = "Not Available"

    try:
        reviewer = review.find_element(By.XPATH, ".//span[contains(@class, 'author')]//span//span").text
    except:
        reviewer = "Anonymous"

    reviews.append({
        "Title": title,
        "Review Text": text,
        "Date": date,
        "Rating": rating,
        "Reviewer Name": reviewer,
        "Source": "BestBuy Canada"
    })

# Convert to Pandas DataFrame
df = pd.DataFrame(reviews)

# Save to CSV
df.to_csv("bestbuy_reviews.csv", index=False, encoding="utf-8")

# Print sample output
print(df.head())

# Close WebDriver
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait, Select
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time
import re
from datetime import datetime
from collections import defaultdict

# Set the path to chromedriver
CHROMEDRIVER_PATH = r"C:\Users\Geetisha Mishra\Documents\Projects\Axion Ray\chromedriver.exe"

# Setup WebDriver
service = Service(CHROMEDRIVER_PATH)
driver = webdriver.Chrome(service=service)

# Open the BestBuy reviews page
url = "https://www.bestbuy.ca/en-ca/product/microsoft-surface-laptop-13-8-touchscreen-copilot-pc-laptop-snapdragon-x-plus-16gb-ram-512gb-ssd-2024-black-en/17924130/review"
driver.get(url)

# Wait until the reviews load
wait = WebDriverWait(driver, 15)

# Filters to apply (prioritized order)
priority_filters = [
    "Most Helpful",
    "Newest",
    "Highest Rating",
    "Lowest Rating",
    "Most Relevant",
    "Oldest"
]

filter_values = {
    "Most Relevant": "relevancy",
    "Most Helpful": "helpfulness",
    "Newest": "newest",
    "Oldest": "oldest",
    "Highest Rating": "highestRating",
    "Lowest Rating": "lowestRating"
}

# Dictionary to store unique reviews with assigned filters
reviews_dict = {}

# Dictionary to track review count per filter
filter_counts = defaultdict(int)

# Target distribution (example: adjust these values based on actual needs)
target_distribution = {
    "Most Helpful": 10,
    "Newest": 10,
    "Highest Rating": 8,
    "Lowest Rating": 8,
    "Most Relevant": 12,
    "Oldest": 10
}

# Function to apply a filter and scrape reviews
def scrape_reviews(filter_name, filter_value, max_reviews):
    print(f"\nApplying filter: {filter_name}")

    # Locate the dropdown and select the filter
    select_element = wait.until(EC.presence_of_element_located((By.ID, "Sort")))
    select = Select(select_element)
    select.select_by_value(filter_value)
    time.sleep(3)  # Wait for the page to reload with new reviews

    # Load all reviews by clicking "Show More"
    while True:
        try:
            show_more_button = wait.until(EC.element_to_be_clickable((By.XPATH, "//button[contains(text(), 'Show More')]")))
            driver.execute_script("arguments[0].click();", show_more_button)
            time.sleep(2)  # Allow time for new reviews to load
        except:
            break  # No more "Show More" button, exit loop

    # Extract all reviews
    review_elements = driver.find_elements(By.XPATH, "//li[contains(@class, 'review')]")
    print(f"Total reviews found for {filter_name}: {len(review_elements)}")

    for review in review_elements:
        if filter_counts[filter_name] >= max_reviews:
            break  # Stop when the required count is reached

        try:
            title = review.find_element(By.XPATH, ".//div[contains(@class, 'reviewTitle')]//span").text.strip()
        except:
            title = "Not Available"

        try:
            text = review.find_element(By.XPATH, ".//div[contains(@class, 'reviewContent')]").text.strip()
        except:
            text = "Not Available"

        try:
            date_raw = review.find_element(By.XPATH, ".//span[contains(@class, 'locationAndTime')]").text.strip()
            try:
                date = datetime.strptime(date_raw.replace("-", ""), "%B %d, %Y").strftime("%Y-%m-%d")
            except:
                date = date_raw  # Keep as is if parsing fails
        except:
            date = "Not Available"

        try:
            rating_text = review.find_element(By.XPATH, ".//p[contains(text(), 'Customer rating')]").text
            rating = rating_text.split(":")[1].split(" out of")[0].strip()
        except:
            rating = "Not Available"

        try:
            reviewer = review.find_element(By.XPATH, ".//span[contains(@class, 'author')]//span//span").text.strip()
        except:
            reviewer = "Anonymous"

        # Generate a unique key using Review Text (normalized)
        review_key = re.sub(r'\W+', '', text.lower())  # Remove special characters

        if review_key not in reviews_dict:
            # If review is new, add it with its filter
            reviews_dict[review_key] = {
                "Sort Filter": filter_name,  # Assign the FIRST filter seen
                "Title": title,
                "Review Text": text,
                "Date": date,
                "Rating": rating,
                "Reviewer Name": reviewer,
                "Source": "BestBuy Canada"
            }
            filter_counts[filter_name] += 1  # Increment count for this filter

# Apply filters while ensuring balanced distribution
for filter_name in priority_filters:
    scrape_reviews(filter_name, filter_values[filter_name], target_distribution[filter_name])

# Close WebDriver
driver.quit()

# Convert to Pandas DataFrame
df = pd.DataFrame(reviews_dict.values())

# Ensure exactly 58 reviews
if len(df) > 58:
    df = df.sample(n=58, random_state=42)  # Trim down to 58 if extra
elif len(df) < 58:
    print(f"‚ö†Ô∏è Warning: Only {len(df)} reviews found. Needed 58.")

# Add a Primary Key (Unique ID)
df.insert(0, "Review_ID", ["R" + str(i).zfill(6) for i in range(1, len(df) + 1)])

# Save the final dataset as a single CSV
df.to_csv("bestbuy_reviews_balanced.csv", index=False, encoding="utf-8")

# Print sample output
print(df.head())
print(f"‚úÖ Total unique reviews: {len(df)} (should be exactly 58)")
print(f"üìÅ Balanced CSV saved as 'bestbuy_reviews_balanced.csv'")
```

```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Load the cleaned dataset
df = pd.read_csv("bestbuy_reviews_step3_missing_values_fixed.csv")

# Initialize VADER Sentiment Analyzer
analyzer = SentimentIntensityAnalyzer()

# Function to classify sentiment
def get_sentiment(text):
    scores = analyzer.polarity_scores(text)
    if scores["compound"] >= 0.05:
        return "Positive"
    elif scores["compound"] <= -0.05:
        return "Negative"
    else:
        return "Neutral"

# Apply sentiment analysis to Review Text
df["Sentiment"] = df["Review Text"].apply(get_sentiment)

# Categorize sentiment labels
def categorize_sentiment(sentiment):
    if sentiment == "Positive":
        return "['Good Design & Quality (Pos)']"
    elif sentiment == "Neutral":
        return "['Satisfactory & Recommended (Neutral)']"
    else:
        return "['Product Issues (Neg)']"

# Assign categorized sentiment labels
df["Sentiment Category"] = df["Sentiment"].apply(categorize_sentiment)

# Save results
df.to_csv("bestbuy_reviews_with_sentiment.csv", index=False)

# Display sample results
df[["Review_ID", "Title", "Review Text", "Sentiment", "Sentiment Category"]].head(10)
```
